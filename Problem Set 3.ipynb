```python
import pandas as pd
import numpy as np
dataset_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users_data = pd.read_csv(dataset_url, sep='|')
users_data = pd.read_csv(dataset_url, sep='|')  # Assuming '|' is the delimiter
mean_age_by_occupation = users_data.groupby('occupation')['age'].mean()
print(mean_age_by_occupation)
mean_age_by_occupation = users_data.groupby('occupation')['age'].mean()
print(mean_age_by_occupation)
# Total number of users per occupation
total_users_by_occupation = users_data.groupby('occupation')['gender'].count()

# Total number of males per occupation
total_males_by_occupation = users_data[users_data['gender'] == 'M'].groupby('occupation')['gender'].count()

# Male ratio per occupation
male_ratio_by_occupation = (total_males_by_occupation / total_users_by_occupation).sort_values(ascending=False)
print(male_ratio_by_occupation)
min_max_age_by_occupation = users_data.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max_age_by_occupation)
mean_age_by_occupation_and_gender = users_data.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_by_occupation_and_gender)
# Gender counts per occupation
gender_counts_by_occupation = users_data.groupby(['occupation', 'gender']).size().unstack()

# Gender percentage per occupation
gender_percentage_by_occupation = (gender_counts_by_occupation.div(gender_counts_by_occupation.sum(axis=1), axis=0) * 100)
print(gender_percentage_by_occupation)

```

```python
# Question 4
Wind Statistics
The data have been modified to contain some missing values, identified by NaN.
Using pandas should make this exercise easier, in particular for the bonus question.
You should be able to perform all of these operations without using a for loop or
other looping construct.
The data in 'wind.data' has the following format:
Yr Mo Dy RPT VAL ROS KIL SHA BIR DUB CLA MUL CLO BEL
MAL
61 1 1 15.04 14.96 13.17 9.29 NaN 9.87 13.67 10.25 10.83 12.58 18.50 15.04
61 1 2 14.71 NaN 10.83 6.50 12.62 7.67 11.50 10.04 9.79 9.67 17.54 13.83
61 1 3 18.50 16.88 12.33 10.13 11.17 6.17 11.25 NaN 8.50 7.67 12.75 12.71

The first three columns are year, month, and day. The remaining 12 columns are
average windspeeds in knots at 12 locations in Ireland on that day.
Step 1. Import the necessary libraries
Step 2. Import the dataset from the attached file wind.txt
Step 3. Assign it to a variable called data and replace the first 3 columns by a proper
datetime index.
Step 4. Year 2061? Do we really have data from this year? Create a function to fix it
and apply it.
Step 5. Set the right dates as the index. Pay attention at the data type, it should be
datetime64[ns].
Step 6. Compute how many values are missing for each location over the entire
record.They should be ignored in all calculations below.
Step 7. Compute how many non-missing values there are in total.
Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and
all the times.
A single number for the entire dataset.
Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean
windspeeds and standard deviations of the windspeeds at each location over all the
days
A different set of numbers for each location.
Step 10. Create a DataFrame called day_stats and calculate the min, max and mean
windspeed and standard deviations of the windspeeds across all the locations at each
day.
A different set of numbers for each day.
Step 11. Find the average windspeed in January for each location.
Treat January 1961 and January 1962 both as January.
Step 12. Downsample the record to a yearly frequency for each location.
Step 13. Downsample the record to a monthly frequency for each location.
Step 14. Downsample the record to a weekly frequency for each location.
Step 15. Calculate the min, max and mean windspeeds and standard deviations of the
windspeeds across all locations for each week (assume that the first week starts on
January 2 1961) for the first 52 weeks.
```

```python
import pandas as pd
import numpy as np
from io import StringIO

# Sample data
data_text = """Yr Mo Dy RPT VAL ROS KIL SHA BIR DUB CLA MUL CLO BEL MAL
61 1 1 15.04 14.96 13.17 9.29 NaN 9.87 13.67 10.25 10.83 12.58 18.50 15.04
61 1 2 14.71 NaN 10.83 6.50 12.62 7.67 11.50 10.04 9.79 9.67 17.54 13.83
61 1 3 18.50 16.88 12.33 10.13 11.17 6.17 11.25 NaN 8.50 7.67 12.75 12.71
"""

# Step 1: Read data
data = pd.read_csv(StringIO(data_text), sep=r'\s+')


# Step 2: Convert Yr, Mo, Dy to datetime and fix the year
data['datetime'] = pd.to_datetime({'year': data['Yr'] + 1900, 'month': data['Mo'], 'day': data['Dy']})
data = data.drop(['Yr', 'Mo', 'Dy'], axis=1).set_index('datetime')

# Step 3: Ensure datetime index
data.index = pd.to_datetime(data.index)

# Step 4: Missing values count
missing_values_per_location = data.isnull().sum()

# Step 5: Non-missing values count
non_missing_values_total = data.notnull().sum().sum()

# Step 6: Mean windspeed over all locations and times
mean_windspeed = data.mean().mean()

# Step 7: Location stats
loc_stats = data.describe().T[['min', 'max', 'mean', 'std']]

# Step 8: Day stats
day_stats = data.aggregate(['min', 'max', 'mean', 'std'], axis=1)

# Step 9: Average windspeed in January
january_data = data[data.index.month == 1]
january_avg_windspeed = january_data.groupby(january_data.index.year).mean()

# Step 10: Downsample to yearly frequency
yearly_downsample = data.resample('YE').mean()

# Step 11: Downsample to monthly frequency
monthly_downsample = data.resample('ME').mean()

# Step 12: Downsample to weekly frequency
weekly_downsample = data.resample('W').mean()

# Step 13: Weekly statistics for first 52 weeks
weekly_stats = data.resample('W').agg(['min', 'max', 'mean', 'std']).iloc[:52]

# Print results
print(f"Missing values per location:\n{missing_values_per_location}\n")
print(f"Non-missing values total: {non_missing_values_total}")
print(f"Mean windspeed: {mean_windspeed}")
print(f"Location statistics:\n{loc_stats}\n")
print(f"Day statistics:\n{day_stats}\n")
print(f"January average windspeed:\n{january_avg_windspeed}\n")
print(f"Yearly downsample:\n{yearly_downsample}\n")
print(f"Monthly downsample:\n{monthly_downsample}\n")
print(f"Weekly downsample:\n{weekly_downsample}\n")
print(f"Weekly stats for first 52 weeks:\n{weekly_stats}\n")
```

```python
# Question 5
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called chipo.
Step 4. See the first 10 entries
Step 5. What is the number of observations in the dataset?
Step 6. What is the number of columns in the dataset?
Step 7. Print the name of all the columns.
Step 8. How is the dataset indexed?
Step 9. Which was the most-ordered item?
Step 10. For the most-ordered item, how many items were ordered?
Step 11. What was the most ordered item in the choice_description column?
Step 12. How many items were orderd in total?
Step 13.
• Turn the item price into a float
• Check the item price type
• Create a lambda function and change the type of item price
• Check the item price type
Step 14. How much was the revenue for the period in the dataset?
Step 15. How many orders were made in the period?
Step 16. What is the average revenue amount per order?
Step 17. How many different items are sold?
```

```python

import pandas as pd
import numpy as np

# Correct URL with proper delimiter
dataset_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(dataset_url, sep='\t')  # Specify tab delimiter
print(chipo.head(10))
num_observations = chipo.shape[0]
print(f"Number of observations: {num_observations}")
num_columns = chipo.shape[1]
print(f"Number of columns: {num_columns}")
column_names = chipo.columns
print(f"Column names: {column_names}")
dataset_index = chipo.index
print(f"Dataset index: {dataset_index}")
most_ordered_item = chipo.groupby('item_name')['quantity'].sum().idxmax()
print(f"Most-ordered item: {most_ordered_item}")
quantity_most_ordered = chipo.groupby('item_name')['quantity'].sum().max()
print(f"Quantity of the most-ordered item: {quantity_most_ordered}")
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()
print(f"Most-ordered item in choice_description: {most_ordered_choice}")
total_items_ordered = chipo['quantity'].sum()
print(f"Total items ordered: {total_items_ordered}")
print(chipo['item_price'].dtype)
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.replace('$', '')))
print(chipo['item_price'].dtype)
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print(f"Revenue: ${revenue:.2f}")
num_orders = chipo['order_id'].nunique()
print(f"Number of orders: {num_orders}")
average_revenue_per_order = revenue / num_orders
print(f"Average revenue per order: ${average_revenue_per_order:.2f}")
num_different_items = chipo['item_name'].nunique()
print(f"Number of different items sold: {num_different_items}")
```

```python
# Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the dataset
data = {
    "Year": [1867, 1900, 1920, 1940, 1960, 1980, 2000, 2014],
    "Marriages": [10, 12, 15, 20, 30, 40, 35, 32],
    "Divorces": [2, 3, 5, 8, 15, 25, 30, 28]
}
df = pd.DataFrame(data)

# Step 2: Create the plot
plt.figure(figsize=(12, 6))

# Plotting the data with unique styles
plt.plot(df['Year'], df['Marriages'], label='Marriages per Capita', marker='o', linestyle='--', color='blue', linewidth=2)
plt.plot(df['Year'], df['Divorces'], label='Divorces per Capita', marker='s', linestyle='-.', color='red', linewidth=2)

# Add annotations to highlight key points
plt.annotate('Peak Marriages', xy=(1960, 30), xytext=(1950, 35),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12)
plt.annotate('Steady Divorces', xy=(1980, 25), xytext=(1990, 20),
             arrowprops=dict(facecolor='gray', arrowstyle='->'), fontsize=12)

# Highlight specific periods with shaded areas
plt.axvspan(1900, 1920, color='lightblue', alpha=0.3, label='Early 20th Century')

# Add title and axis labels with customized styling
plt.title('U.S. Marriages and Divorces per Capita (1867-2014)', fontsize=16, fontweight='bold', color='darkgreen')
plt.xlabel('Year', fontsize=14, fontstyle='italic', color='darkblue')
plt.ylabel('Per Capita Rate', fontsize=14, fontstyle='italic', color='darkblue')

# Customize the legend
plt.legend(fontsize=12, loc='upper left')

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```

```python
# Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the dataset with unique actor names
data = {
    "Actor": ["John Killman", "Max Slaymore", "Rick Deathshot", "Clint Firestorm", "Chuck Blaster"],
    "Kill Count": [500, 400, 350, 320, 290]
}
df = pd.DataFrame(data)

# Step 2: Sort the dataset by kill count
df = df.sort_values(by="Kill Count", ascending=False)

# Step 3: Create the horizontal bar chart
plt.figure(figsize=(10, 6))

# Plotting the horizontal bars
plt.barh(df['Actor'], df['Kill Count'], color='darkblue', edgecolor='black')

# Adding labels and title
plt.title('Deadliest Actors in Hollywood', fontsize=14, fontweight='bold')
plt.xlabel('Kill Count', fontsize=12)
plt.ylabel('Actor', fontsize=12)

# Adding values to the bars
for index, value in enumerate(df['Kill Count']):
    plt.text(value + 10, index, str(value), fontsize=10, va='center')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Question 10
Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the dataset
data = {
    "Year": [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009],
    "Arcade Revenue (in millions)": [1200, 1150, 1100, 1050, 1025, 1000, 950, 900, 850, 800],
    "CS PhDs Awarded": [800, 825, 850, 875, 890, 920, 940, 960, 980, 1000]
}
df = pd.DataFrame(data)

# Step 2: Create the scatter plot
plt.figure(figsize=(10, 6))

# Create scatter plot, coloring points by year
scatter = plt.scatter(
    df["Arcade Revenue (in millions)"], 
    df["CS PhDs Awarded"], 
    c=df["Year"], 
    cmap="viridis", 
    edgecolor="black", 
    s=100
)

# Adding colorbar
colorbar = plt.colorbar(scatter)
colorbar.set_label("Year")

# Add labels and title
plt.title("Relationship Between Arcade Revenue and CS PhDs Awarded (2000-2009)", fontsize=14, fontweight="bold")
plt.xlabel("Arcade Revenue (in millions)", fontsize=12)
plt.ylabel("CS PhDs Awarded", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
